```python
# MOH ISRAN
# F55121073
# B
import time

# Algoritma TSP
def tsp(graph, start):
    num_nodes = len(graph)
    unvisited = set(range(num_nodes))
    unvisited.remove(start)
    path = [start]
    current_node = start
    total_distance = 0

    while unvisited:
        next_node = None
        min_distance = float('inf')
        for node in unvisited:
            if graph[current_node][node] < min_distance:
                min_distance = graph[current_node][node]
                next_node = node

        unvisited.remove(next_node)
        path.append(next_node)
        total_distance += min_distance
        current_node = next_node
        print(f"Iterasi: {path}, Distance: {total_distance}")

    path.append(start)
    total_distance += graph[current_node][start]
    print(f"Shortest Path: {path}")
    print(f"Total Distance: {total_distance}")
    return path, total_distance


# Algoritma Dijkstra
def dijkstra(graph, start):
    num_nodes = len(graph)
    distance = [float('inf')] * num_nodes
    distance[start] = 0
    path = [start]
    current_node = start
    visited = set()

    while len(visited) < num_nodes:
        visited.add(current_node)
        next_node = None
        min_distance = float('inf')

        for node in range(num_nodes):
            if node not in visited and graph[current_node][node] > 0:
                new_distance = distance[current_node] + graph[current_node][node]
                if new_distance < distance[node]:
                    distance[node] = new_distance
                    if new_distance < min_distance:
                        min_distance = new_distance
                        next_node = node

        if next_node is None:
            unvisited = set(range(num_nodes)) - visited
            next_node = min(unvisited, key=lambda x: distance[x])

        path.append(next_node)
        current_node = next_node
        print(f"Iterasi: {path}, Distance: {distance[current_node]}")

    print(f"Shortest Path: {path}")
    print(f"Total Distance: {distance[current_node]}")
    return path, distance[current_node]


# Fungsi untuk menghitung waktu komputasi
def compute_time(func, *args):
    start_time = time.time()
    result = func(*args)
    end_time = time.time()
    execution_time = end_time - start_time
    print(f"Execution Time: {execution_time} seconds")
    return result


# Fungsi utama
def main():
    # Contoh graph (adjacency matrix)
    graph_tsp = [
        [0, 2, 9, 10],
        [1, 0, 6, 4],
        [15, 7, 0, 8],
        [6, 3, 12, 0]
    ]

    graph_dijkstra = [
        [0, 4, 0, 0, 0, 0, 0, 8, 0],
        [4, 0, 8, 0, 0, 0, 0, 11, 0],
        [0, 8, 0, 7, 0, 4, 0, 0, 2],
        [0, 0, 7, 0, 9, 14, 0, 0, 0],
        [0, 0, 0, 9, 0, 10, 0, 0, 0],
        [0, 0, 4, 14, 10, 0, 2, 0, 0],
        [0, 0, 0, 0, 0, 2, 0, 1, 6],
        [8, 11, 0, 0, 0, 0, 1, 0, 7],
        [0, 0, 2, 0, 0, 0, 6, 7, 0]
    ]

    print("Pilih algoritma:")
    print("1. TSP")
    print("2. Dijkstra")
    choice = int(input("Masukkan pilihan (1/2): "))

    if choice == 1:
        start_node = int(input("Masukkan node awal: "))
        compute_time(tsp, graph_tsp, start_node)
    elif choice == 2:
        start_node = int(input("Masukkan node awal: "))
        compute_time(dijkstra, graph_dijkstra, start_node)
    else:
        print("Pilihan tidak valid.")


if __name__ == '__main__':
    main()
```
